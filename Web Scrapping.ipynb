```python
#Web Scrapping - extracting data from web
#library used - Beautiful soup
import requests
from bs4 import BeautifulSoup as bs
url = 'https://www.amazon.in/Apple-iPhone-Pro-Max-256GB/product-reviews/B09G91FCBT/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page = requests.get(url)
page
```

```python
page.content #we get the whole source code of the amazon customer reviws page
```

```python
soup = bs(page.content,'html.parser')
soup #html.parser is a better readable format of the customer review page source code
```

```python
#never ever consider the top positive or top critical review
names = soup.find_all('span',class_ = 'a-profile-name')[2:]
names
```

```python
len(names)
```

```python
import pandas as pd
df = pd.DataFrame(names,columns = {'Customer Name'})
df
```

```python
#in the data frame we have few duplicates , we have to drop the duplicates
df = df.drop_duplicates(ignore_index = True) #
df
```

```python
#now the nect item is review title
r_title = soup.find_all('a',class_ = 'review-title')
r_title
```

```python
len(r_title)
```

```python
review_title = []  #empty list
for i in range(0,len(r_title)):
    review_title.append(r_title[i].get_text()[1:-1])
review_title
```

```python
#The next element is review date
dates = soup.find_all('span',class_ = 'review-date')[2:]
dates
```

```python
len(dates)
```

```python
review_date = []   #emptylist
for i in range(0,len(dates)):
    review_date.append((dates[i].get_text().replace('Reviewed in India on ','')))
review_date
```

```python
#Next element is review rating in stars
rating = soup.select('i.review-rating')[2:]
rating
```

```python
len(rating)
```

```python
review_rating = []   #emptylist
for i in range(0,len(rating)):
    review_rating.append(rating[i].get_text())
review_rating
```

```python
#review content
content = soup.select('span.review-text-content')
content
```

```python
review_content = []
for i in range(0,len(content)):
    review_content.append(content[i].get_text().replace('\n',''))
review_content
```

```python
df
```

```python
df['Review Title'] = review_title
df['Review Date'] = review_date
df['Review Content'] = review_content
df['Review Stars'] = review_rating
df
```

```python
df.to_csv('Amazon_Reviews.csv')
```
